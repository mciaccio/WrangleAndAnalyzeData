```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open('merge2.pkl', 'rb') as f:
    merge2_df = pickle.load(f)
print('merge2_df.shape')
print(merge2_df.shape)
```

```python
# populate tweet_hour pandas Series
# required for visualization 
merge2_df.tweet_hour = merge2_df.timestamp.apply(lambda x: x.hour)
print('merge2_df.tweet_hour')
print(merge2_df.tweet_hour.head(2))
```

```python
# Create plot for act_report.pdf
# tweet frequency by hour of day
plt.rcParams.update({'font.size': 15})
plot = merge2_df.tweet_hour.value_counts()[:6].plot(kind='bar', 
             title='tweet frequency by hour', legend=False)

plot.text(0.85, 0.85, '0 hour\nmidnight', horizontalalignment='center',
     verticalalignment='center', transform=plot.transAxes,
         fontsize='x-small', 
           bbox=dict(facecolor='none', boxstyle='round'))

plt.ylabel('Count')
plt.xlabel('Hour')

# plt.show()
fig = plot.get_figure()
fig.savefig('tweet_by_hour', bbox_inches = 'tight')
```

```python
# Create plot for act_report.pdf
plt.rcParams.update({'font.size': 15})
plot = merge2_df.p1.value_counts()[:6].plot(kind='bar', 
             title='Most Likely Neural Net Breed Prediction Count')

plt.ylabel('Count')
plt.xlabel('Breed')

# plt.show()
fig = plot.get_figure()
fig.savefig('1_Neural_Net_Pred.png', bbox_inches = 'tight')
```
